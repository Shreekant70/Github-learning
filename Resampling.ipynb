```python
from sklearn.datasets import make_classification
from sklearn.utils import resample
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data = make_classification()
```

```python
data = make_classification(n_samples= 1000 , n_features =10, n_classes =2 ,weights=[0.8,0.2])
```

```python
X =data[0]
y=data[1]
```

```python
df =pd.DataFrame(X ,columns =["col_{}".format(i+1) for i in range(len(X[0]))])
```

```python
df['target']  = y
```

```python
# Down -sampling ----> Delecting rows from big class
# Data became less

# Up-sampling----> Duplicating Random rows from small class 
# Data beame double
```

```python
zero =df[df["target"]==0]
one = df[df["target"]==1]
```

```python
down = resample(zero, replace=False,n_samples=len(one))
```

```python
down
```

```python
len(one)
```

```python
down_samples=pd.concat([one,down], axis=0)
```

```python
down_samples
```

```python
up = resample(one, replace=True,n_samples=len(zero))
```

```python
up
```

```python
up_samples=pd.concat([zero,up], axis=0)
```
